# Parámetros T - Ejemplos iniciales

## Repaso

Una red caracterizada por sus parámetros T la podemos plantear a partir de:

$$
\begin{equation*}
\mathbf{T} =  \begin{bmatrix}
  A & B \\
  C & D \\
  \end{bmatrix}
\end{equation*}
$$

Donde:

$A = \frac{V_1}{V_2} \mapsto$ con $I_2 = 0$ 

$B = \frac{V_1}{-I_2} \mapsto$ con $V_2 = 0$

$C = \frac{I_1}{V_2} \mapsto$ con $I_2 = 0$

$D = \frac{I_1}{-I_2} \mapsto$ con $V_2 = 0$


**Condiciones de medición**

- Cuando $I_2 = 0$, cortocircuitamos la salida.

- Cuando $V_2 = 0$, dejamos la salida en circuito abierto.

## Impedancia en serie

<img src='./imagenes/z.jpeg' width="300">

Analicemos esta celda fundamental:

$A = \frac{V_1}{V_2} \mapsto$ para $I_2 = 0$

Dado que $I_2 = 0$, entonces $I_1 = 0$, y $V_1 = V_2$

$\mapsto A = \frac{V1}{V2} = 1$

$B = \frac{V_1}{-I_2} \mapsto$ para $V_2 = 0$

$V_1 = I_1 Z$

$-I_2 = I_1$

$\mapsto B = \frac{V_1}{-I_2} = Z$

$C = \frac{I_1}{V_2} \mapsto$ para $I_2 = 0$

Siendo $I_2 = 0$, $I_1 = 0$

$\mapsto C = 0$

$D = \frac{I_1}{-I_2} \mapsto$ para $V_2 = 0$


Como analizamos previamente, $I_1 = -I_2$, entonces:

$\mapsto D = 1$

Resumiendo:

$$
\begin{equation*}
\mathbf{T} =  \begin{bmatrix}
  1 & Z \\
  0 & 1 \\
  \end{bmatrix}
\end{equation*}
$$

## Impedancia en paralelo

<img src='./imagenes/y.jpeg' width="300">

Analicemos esta celda fundamental:

$A = \frac{V_1}{V_2} \mapsto$ para $I_2 = 0$

Dado que $I_2 = 0$, entonces $V_1 = V_2$

$\mapsto A = \frac{V1}{V2} = 1$

$B = \frac{V_1}{-I_2} \mapsto$ para $V_2 = 0 = V_1$ 

$\mapsto B = 0$

$C = \frac{I_1}{V_2} \mapsto$ para $I_2 = 0$

Siendo $I_1 = V_1$ y $V_1 = V_2$

$\mapsto C = Y$

$D = \frac{I_1}{-I_2} \mapsto$ para $V_2 = 0$

Como analizamos previamente, $I_1 = -I_2$, entonces:

$\mapsto D = 1$

Resumiendo:

$$
\begin{equation*}
\mathbf{T} =  \begin{bmatrix}
  1 & 0 \\
  Y & 1 \\
  \end{bmatrix}
\end{equation*}
$$

## Circuito L (Z - Y)

<img src='./imagenes/l.jpeg' width="300">

Analicemos esta celda fundamental:

$A = \frac{V_1}{V_2} \mapsto$ para $I_2 = 0$

Dado que $I_2 = 0$, entonces:

$V_2 = V_1 \frac{\frac{1}{Y}}{Z + \frac{1}{Y}} = V_1 \frac{1}{ZY + 1}$

$\mapsto A = \frac{V1}{V2} = 1 + ZY$

$B = \frac{V_1}{-I_2} \mapsto$ para $V_2 = 0$ 

$I_1 = -I_2 = \frac{V_1}{Z}$

$\mapsto B = Z$

$C = \frac{I_1}{V_2} \mapsto$ para $I_2 = 0$

$I_1 = V_2 Y$

$\mapsto C = Y$

$D = \frac{I_1}{-I_2} \mapsto$ para $V_2 = 0$

Como analizamos previamente, $I_1 = -I_2$, entonces:

$\mapsto D = 1$

Resumiendo:

$$
\begin{equation*}
\mathbf{T} =  \begin{bmatrix}
  1 + ZY & Z \\
  Y & 1 \\
  \end{bmatrix}
\end{equation*}
$$

## Circuito T

<img src='./imagenes/t.jpeg' width="300">

Analicemos esta celda fundamental:

$A = \frac{V_1}{V_2} \mapsto$ $I_2 = 0$

$\frac{1}{A} = \frac{Z_2}{Z_1 + Z_2}$

$\mapsto A = 1 + Z_1 Y_2$

$B = \frac{V_1}{-I_2} \mapsto$ para $V_2 = 0$ 

$V_x = V_1 \frac{\frac{1}{Y_2 + Y_3}}{\frac{1}{Y_2 + Y_3} + Z_1} $

$V_x = V_1 \frac{1}{1 + Z_1 (Y_2 + Y_3)}$

$-I_2 = \frac{V_x}{Z_3}$

$-I_2 = V_1 \frac{1}{Z_3[1 + Z_1 (Y_2 + Y_3)]}$

$\mapsto B = Z_3[1 + Z_1 (Y_2 + Y_3)]$

$C = \frac{I_1}{V_2} \mapsto$ para $I_2 = 0$

$\mapsto C = Y_2$

$D = \frac{I_1}{-I_2} \mapsto$ para $V_2 = 0$

$\frac{1}{D} = \frac{Z_2}{Z_2 + Z_3}$

$\mapsto D = 1 + Y_2 Z_3$

$$
\begin{equation*}
\mathbf{T} =  \begin{bmatrix}
  1 + Z_1 Y_2 & Z_3[1 + Z_1(Y_2 + Y_3)] \\
  Y_2 & 1 + Y_2 Z_3 \\
  \end{bmatrix}
\end{equation*}
$$
