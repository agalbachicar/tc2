## Análisis

El siguiente circuito puede resultar un poco complicado de analizar la primera vez que se lo trata. El siguiente desarrollo, pretende guiar al estudiante en la metodología para poder analizar la impedancia de entrada del siguiente circuito.

<img src="./conversor_impedancia/conversor.png">

Nos interesa obtener la expresión de la impedancia de entrada, la cual la denominaremos $Z_{in} = \frac{V_1}{I_1}$. Para resolver el circuito, utilizaremos expresiones genéricas de impedancia, denominadas $Z_i$ tal que $_i$ será el número de impedancia que se relaciona con la figura anterior, al obtener la expresión final, lo reemplazaremos por el valor de la impedancia que corresponde.

Entonces, los terminales de entrada de los amplificadores operacionales $U_1$ y $U_2$. Por condición de amplificador operacional ideal, en particular, la impedancia de entrada tiende a infinito, la tensión diferencial es nula en ambos operacionales tal que:

$V_1 = V_2$

$V_2 = V_3$

Por transitividad:

$V_1 = V_3$

Los terminales $4$ y $5$ son las salidas de los amplificadores, que nuevamente por condición de amplificador operacional ideal (característica de salida, generador ideal de tensión) será fijada por circuito externo y suministrará la corriente necesaria para mantener ese nivel de tensión. La corriente desde la salida de los operacionales a los terminales $4$ y $5$ es desconocida por lo que no nos servirá en nuestro análisis.

Lo que si podemos utilizar son las tensiones $V_4$ y $V_5$ tal que:

$I_1 = \frac{V_1 - V_4}{Z_1}$

$I_2 = \frac{V_2 - V_4}{Z_2}$

$I_3 = \frac{V_5 - V_2}{Z_3}$

$I_4 = \frac{V_5 - V_3}{Z_4}$

$I_5 = \frac{V_3}{Z_5}$

A nivel corrientes, sabemos que:

$I_2 = I_3$

$I_4 = I_5$

Así, reemplzamos todas las ocurrencias de $V_2$ y $V_3$ por $V_1$:

$I_1 = \frac{V_1 - V_4}{Z_1} (1)$

$I_2 = \frac{V_1 - V_4}{Z_2} (2)$

$I_3 = \frac{V_5 - V_1}{Z_3} (3)$

$I_4 = \frac{V_5 - V_1}{Z_4} (4)$

$I_5 = \frac{V_1}{Z_5} (5)$

$I_2 = I_3 (6)$ 

$I_4 = I_5 (7)$

Usamos $(7)$ para despejar $V_5$ de $(5)$ y $(4)$ y reemplazamos en $(3)$:

$-\frac{V_1}{Z_5} = \frac{V_5 - V_1}{Z_4}$

$ V_1 Z_4(\frac{1}{Z_4} + \frac{1}{Z_5}) = V_5 $

$I_3 = V_1 \frac{Z_4}{Z_3}(\frac{1}{Z_4} + \frac{1}{Z_5}) - V_1 \frac{1}{Z_3} $

$I_3 = V_1 \frac{Z_4 + Z_5}{Z_3 Z_5} - V_1 \frac{1}{Z_3} $

$I_3 = V_1 \frac{Z_4}{Z_3 Z_5} (8)$
 
Usamos $(8)$ y $(6)$ con $(2)$ y despejamos $V_4$:

$V_1 \frac{Z_4}{Z_3 Z_5} = \frac{V_1 - V_4}{Z_2} $

$V_1 Z_2 (-\frac{Z_4}{Z_3 Z_5} + \frac{1}{Z_2}) = V_4 $

$V_1 (1 - \frac{Z_4 Z_2}{Z_3 Z_5}) = V_4 $

$V_1 \frac{Z_3 Z_5 -Z_4 Z_2}{Z_3 Z_5} = V_4 (9)$

Finalmente, reemplazamos $(9)$ en $(1)$ y computamos la impedancia de entrada:

$I_1 = \frac{V_1}{Z_1} - V_1 \frac{Z_3 Z_5 - Z_4 Z_2}{Z_1 Z_3 Z_5}$

$I_1 = V_1 \frac{Z_3 Z_5 - Z_3 Z_5 + Z_4 Z_2}{Z_1 Z_3 Z_5}$

$I_1 = V_1 \frac{Z_4 Z_2}{Z_1 Z_3 Z_5}$

$Z_{in} = \frac{V_1}{I_1} = \frac{Z_1 Z_3 Z_5}{Z_4 Z_2} $

$Z_{in} = \frac{Z_1 Z_3 Z_5}{Z_4 Z_2} $

Vemos que la última expresión permite expresar la impedancia de entrada $Z_in$ como el producto y el cociente de las impedancias que componen el circuito. Esto nos permite realizar vairas combinaciones de gran utilidad.

## Circuitos de aplicación

### 1- Inductor equivalente

Este circuito es bastante conveniente debido que puede conformarse dipolos que se comportan como inductores, y de gran tamaño (medido en Hy) con resistencias y capacitores. Además, no se verán generados campos magnéticos asociados al inductor equivalente (puede ser una desventaja según la aplicación). Considere el siguiente arreglo de componentes

- $Z_1$, $Z_2$, $Z_3$ y $Z_5$ son resistencias.
- $Z_4$ es un capacitor.

La impedancia de entrada será:

$Z_{in} = \frac{Z_1 Z_3 Z_5}{Z_4 Z_2} = \frac{R_1 R_3 R_5}{R_2 \frac{1}{sC_4}}$

$Z_{in} = s \frac{R_1 R_3 R_5 C_4}{R_2}$, $L_{eq} = \frac{R_1 R_3 R_5 C_4}{R_2}$

### 2- Resistencia negativa dependiente de la frecuencia (Frequency Dependent Negative Resistor)

Este circuito es comunmente conocido por su acrónimo en inglés, FDNR, y permite generar una resistencia negativa dependiente de la frecuencia. Curiosamente, este circuito genera una relación cuadrática con la frecuencia, a diferencia de los componentes R, L y C que poseen dependencias lineales (directas o inversas) con la frecuencia. Para lograrlo, considere la siguiente atribución de componentes:

- $Z_1$,  $Z_3$ y $Z_5$ son resistencias.
- $Z_2$ y $Z_4$ son capacitores.


La impedancia de entrada será:

$Z_{in} = \frac{Z_1 Z_3 Z_5}{Z_4 Z_2} = \frac{R_1 R_3 R_5}{\frac{1}{sC_2} \frac{1}{sC_4}}$

$Z_{in} = s^2 R_1 R_3 R_5 C_2 C_4 $

$Z_{{in}_{(jw)}} = - w^2 R_1 R_3 R_5 C_2 C_4 $

**Tarea**

### 3- Componentes escalados, integración

Como vemos, tanto en numerador como en denominador se dispone de una multiplicatoria de componentes cuyos coeficientes pueden ser escalados en orden de magnitud rápidamente sin necesidad de variar su tamaño ni tecnología. Esto lo vuelve especialemente atractivo para generar circuitos integrados que permitan integrar grandes valores de capacidad, inductancia o resistencia.

## Consideraciones del modelo

Es preciso notar que el modelo circuital plantea amplificadores ideales los cuales presentan algunas diferencias respecto a los reales, a saber y tener en cuenta al momento de implementar:

- Las tensiones de modo común: las entradas del amplificador deben estar sometidas a niveles que no superen los extremos de modo común de los amplificadores operacionales sino corre riesgo que la polarización de los mismos se mueva de rango y el circuito no se comporte como se espera.
- El PGB (Producto Ganancia Ancho de Banda) de los amplificadores juega un papel fundamental debido a las posibles oscilaciones y los rangos de operación de este circuito. En otras palabras, difícilmente podremos con amplificadores convencionales trabajar a frecuencias del orden de los 100KHz o más, incluso bastante menos.
- Los modelos de los componentes no son ideales, por lo que siempre habrá componentes parásitos.
