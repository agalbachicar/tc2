Descripcion: muestra el uso de las funciones disponibles para trabajar con Chebyshev  

```python
import scipy.signal as sig
import matplotlib.pyplot as plt
from splane import pzmap, grpDelay, bodePlot, convert2SOS
```

```python
plt.close('all')
```

```python
ORDER = 3
RIPPLE = 2
```

```python
z,p,k = sig.cheb1ap(ORDER, RIPPLE)
num,den = sig.zpk2tf(z,p,k)
myFilter = sig.TransferFunction(num,den)
```

```python
print(myFilter)
print('Poles:',myFilter.poles)
print('Zeros:',myFilter.zeros)
```

```python
bodePlot(myFilter)
```

```python
pzmap(myFilter) #S plane pole/zero plot
```

```python
grpDelay(myFilter)
```

```python
convert2SOS(myFilter)
```
