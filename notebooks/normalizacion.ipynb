## Conceptos principales

El proceso de normalización (y desnormalización) es una transformación lineal que se aplica una función de red mediante la cual se escala la frecuencia o la impedancia.

Este procedimiento es de suma conveniencia en todo el proceso de diseño de filtros. En otras palabras, lo veremos involucrado en prácticamente todos los procesos de diseño a lo largo del curso debido a lo útil y conveniente que resulta el método para simplificar cálculos y ajustar las soluciones a las restricciones de los problemas.

La selección de las normas o factores de escala es arbitraria, es decir, es responsabilidad del diseñador elegir un factor de escala.

Un detalle no menor, la norma de frecuencia (angular) $\Omega_N$ se mide en $\frac{rad}{s}$. En otras palabras, podemos definir una $f_N$ tal que $\Omega_N = 2 \pi f_N$.


## Tabla de conversión

La siguiente tabla se presenta a modo de resumen. 


| Componente   | Valor desnormalizado | Valor normalizado          |
|--------------|----------------------|----------------------------|
| Resistencia  |          R           |  $\frac{R}{Z_N}$           |
| Inductor     |          L           |  $\frac{L \Omega_N}{Z_N}$  |
| Capacitor    |          C           |  $C \Omega_N Z_N$          |

Donde:

- $Z_N$ es la norma de impedancia.
- $\Omega_N$ es la norma de frecuencia.

Se delega en los alumnos la revisión en la bibliografía el método para derivar en dicha tabla.

## Ejemplos

### Divisor resistivo - transferencia de tensión e impedancia de entrada

Considere el siguiente circuito y calculemos su transferencia de tensión entre los puertos 2 y 1 al ser excitado con un generador de tensión ideal y cargado en vacío.

<img src="./normalizacion/base_r_divider.png">

Así, resulta:

$H_{21} = \frac{V_{R_2}}{V_{R_1}}$

$H_{21} = \frac{R_2}{R_1 + R_2}$

$H_{21} = R_2 \frac{1}{1 + \frac{R_1}{R_2}} (1) $

Analicemos entonces su impedancia de entrada, dada por la función de red de excitación $Z_1 = \frac{V_1}{I_1}$:

$Z_1 = R_1 + R_2 $

$Z_1 = R_2 (1 + \frac{R_1}{R_2}) (2)$

Consideremos la ecuación (1). Vemos que la transferencia de tensión depende de la relación de las impedancias. Consideremos entonces la ecuación (2). Podemos observar que la impedancia de entrada se computa como un factor que escala la resistencia $R_2$ que depende de ambas resistencias.

Este circuito presenta dos componentes, con dos características (trasnferencia de tensión e impedancia de entrada) por lo que podemos ajustar con dichos componentes a tales características. Dado el requisito de transferencia de tensión, y de impedancia de entrada podremos fijar la relación de resistencias para obtener $H_{21}$ y luego el valor de $R_2$ para obtener $Z_1$. Absolutamente quedará definido el valor de $R_1$.

Tomemos el siguiente par de restricciones:

$H_{21} = \frac{1}{3}$

$Z_{1} = 100\Omega$

Podemos despejar:

$R_2 = 33,3\Omega$

$R_1 = 66,6\Omega$

Queremos entonces normalizar el circuito con norma $R_N = 33,3\Omega$. Utilizando la tabla de convesión anterior, debemos dividir por la norma a las resitencias obteniendo:

$R_{2N} = 1$

$R_{1N} = 2$

Analicemos un instante que sucedión con la impedancia de entrada y con la transferencia de tensión:

$H_{21} = R_2 \frac{1}{1 + \frac{R_1}{R_2}} = 1  \frac{1}{1 + \frac{2}{1}} = \frac{1}{3}$

$Z_1 = R_2 (1 + \frac{R_1}{R_2}) =  1 (1 + \frac{2}{1}) = 3$

Podemos observar que la impedancia de entrada se ha escalado con la norma de impedancia, pero la transferencia de tensión no se vio alterada.

### Circuito RLC paralelo - admitancia de entrada

Consideremos ahora el siguiente circuito normalizado tanto en frecuencia como en impedancia.

<img src="./normalizacion/rlc_paralelo.png">

En este caso vamos a calcular la admitancia de entrada $Y_{1_{(s)}}$ y la transferencia de corriente $H_{(s)}$. A tener en cuenta como premisas del circuito y de las funciones a calcular:

- Se alimenta con un generador ideal de corriente.
- Se carga con una impedancia infinita, o bien se lo carga en vacío.
- La corriente $I_1$ es entrante.
- La corriente $I_2$ circula por $C$ del borne superior al borne inferior.
- $Y_{1_{(s)}} = \frac{I_1}{V_1}$
- $H_{(s)} = \frac{I_2}{I_1}$

Admitancia de entrada:

$Y_{1_{(s)}} = \frac{I_1}{V_1}$

$Y_{1_{(s)}} = G + \frac{1}{sL} + sC$

$Y_{1_{(s)}} = \frac{sGL + 1 + s^2 LC}{sL}$

$Y_{1_{(s)}} = LC \frac{s\frac{G}{C} + \frac{1}{LC} + s^2}{sL}$

$Y_{1_{(s)}} = C \frac{s^2 + s\frac{G}{C} + \frac{1}{LC}}{s} (3)$

La transferencia de corriente (planteada como un divisor de corriente):

$H_{(s)} = \frac{I_2}{I_1}$

$I_1 = V_1  Y_1$

$I_2 = V_1  sC$

$H_{(s)} = \frac{V_1 sC}{V_1 C \frac{s^2 + s\frac{G}{C} + \frac{1}{LC}}{s}}$

$H_{(s)} = \frac{s^2}{s^2 + s\frac{G}{C} + \frac{1}{LC}} (4)$

El lector deberá notar que los anteriores coeficientes son normalizados, y para evitar complejizar la interpretación, se omitieron los subíndices $_N$ en $R, C y L$. Entonces la expresión $(3)$ y $(4)$ nos permiten conocer la evolución de la admitancia de entrada y de transferencia de corriente.

Ahora bien, tendremos en nuestro stock de diseños un circuito de estas características y querremos desnormalizarlo a distintos valores de norma de impedancia y de frecuencia. Para conocer los valores desnormalizados, debemos conocer los valores normalizados. En general estos últimos surgen de un proceso de síntesis como efectuamos en el anterior ejemplo. Para este caso, asumimos que los valores son los presentados a continuación:

| Componente | Valor normalizado |
|------------|-------------------|
|      $R_N$ |         1         |
|     $L_N$  |         2         |
|     $C_N$  |         3         |

Y tenemos un factor de desnormalización de $Z_N = 1K\Omega$ y $\Omega_N = 2 \pi 1KHz$. Averiguemos los valores de los componentes:

$ R_N = \frac{R}{Z_N} ; R = R_N x Z_N = 1 x 1K\Omega = 1K\Omega $ 

$ L_N = \frac{L Z_N}{\Omega_N} ; L = \frac{L_N \Omega_N}{Z_N} = \frac{2 x 2\pi 1KHz}{1K\Omega} = 4\pi Hy $ 

$ C_N = C Z_N \Omega_N ; C = \frac{C_N}{Z_N \Omega_N} = \frac{3}{1K\Omega 2 \pi 1KHz} = 477nF $

Verifique los valores con el archivo [rlc_paralelo.asc](./normalizacion/rlc_paralelo.asc).
